```python
# passo 1: abrir navegador e entrar no site de cotação

from selenium import webdriver

navegador = webdriver.Chrome()
```

```python
# passo 2: importar a tabela

import pandas as pd

tabela = pd.read_excel('commodities.xlsx')
display(tabela)
```

```python
# passo 3: pegar o valor da cotação

# import unicodedata
# produto = unicodedata.normalize("NFKD", produto).encode("ascii", "ignore") # --> lib para encode ascii(sem acentos ou caracteres especiais)



for linha in tabela.index:
    produto = tabela.loc[linha, "Produto"]
    
    
    produto = produto.replace("ó", "o").replace("ã", "a").replace("á", "a").replace(
    "ç", "c").replace("ú", "u").replace("é", "e") # --> sem lib
    
    link = f'https://www.melhorcambio.com/{produto}-hoje'
    print(link)
    navegador.get(link)
    
    preco = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    preco = preco.replace('.', '').replace(',', '.')
    
    print(produto)
    print(preco)


    # passo 4: atualizar tabela

    tabela.loc[linha, 'Preço Atual'] = float(preco)


display(tabela)
```

```python
# passo 5: preencher o campo comprar e exportar tabela para pc

tabela['Comprar'] = tabela['Preço Atual'] < tabela['Preço Ideal']
display(tabela)

tabela.to_excel('commodities_atualizado.xlsx', index=False)
navegador.quit()
```
